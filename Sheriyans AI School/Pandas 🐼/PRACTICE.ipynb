```python
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Value': [10, 25, 12, 30, 28, 15, 33, 22],
    'Group_ID': [1, 2, 1, 3, 2, 1, 3, 2]
}

df = pd.DataFrame(data)
df
```

```python
df = df.groupby('Category')['Value'].mean()
df
```

```python
# Example DataFrames for merging and concatenation
import pandas as pd

# First DataFrame
df_a = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df_a
```

```python
# Second DataFrame (with a common 'ID' column)
df_b = pd.DataFrame({
    'ID': [2, 3, 4],
    'Score': [85, 90, 95]
})

df_b
```

```python
concatation = pd.concat([df_a,df_b],axis=1)
concatation
```

```python
concatation2 = pd.concat([df_a,df_b],axis=0)
concatation2
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='outer')
merging
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='inner')
merging
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='left')
merging
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='right')
merging
```
