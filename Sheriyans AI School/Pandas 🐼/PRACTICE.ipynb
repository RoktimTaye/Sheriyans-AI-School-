```python
# Example DataFrames for merging and concatenation
import pandas as pd

# First DataFrame
df_a = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df_a
```

```python
# Second DataFrame (with a common 'ID' column)
df_b = pd.DataFrame({
    'ID': [2, 3, 4],
    'Score': [85, 90, 95]
})

df_b
```

```python
concatation = pd.concat([df_a,df_b],axis=1)
concatation
```

```python
concatation2 = pd.concat([df_a,df_b],axis=0)
concatation2
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='outer')
merging
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='inner')
merging
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='left')
merging
```

```python
merging = pd.merge(df_a, df_b, on='ID', how='right')
merging
```

```python
# Sample DataFrame for pivot table practice
import pandas as pd
import numpy as np

data = {
    'Date': pd.date_range('2023-01-01', periods=8),
    'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Region': ['East', 'West', 'East', 'North', 'West', 'East', 'North', 'West'],
    'Sales': [100, 200, 150, 300, 250, 120, 330, 220],
    'Units': [10, 20, 15, 30, 25, 12, 33, 22]
}

df_pivot = pd.DataFrame(data)
df_pivot
```

```python
# df_original = pd.DataFrame(data)
# pd.pivot_table(df_original, values='Sales', index='Region', columns='Product', aggfunc='median')
pd.pivot_table(df_pivot, values=['Sales','Units'], index='Date', columns='Product', aggfunc='mean')
```

```python
pd.crosstab(df_pivot['Region'], df_pivot['Product'], values=df_pivot['Sales'], aggfunc='sum')
```
