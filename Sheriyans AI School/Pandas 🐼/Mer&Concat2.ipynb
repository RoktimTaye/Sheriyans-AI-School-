```python
import numpy as np 
import pandas as pd 
# DataFrame 1: Employee information

employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how ='inner') # Only keeps the Common data between the two data frames
```

```python
pd.merge(employees,salaries,on='employee_id',how ='outer') # keeps both the data frames details 
```

```python
pd.merge(employees,salaries,on='employee_id',how ='left') # Only keeps the data related to employees data frame
```

```python
pd.merge(employees,salaries,on='employee_id',how ='right') # Only keeps the data related to salaries data frame
```
