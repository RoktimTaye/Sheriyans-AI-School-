```python
import numpy as np 
import pandas as pd
```

```python
# Note: Number of rows should be same tp each other to make a DataFrame perfectly without NaN values or errors.
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] *5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values = "Sales",index = 'Region',columns="Product",aggfunc = 'median') # product ka konsa value(A,B,C,D) kis region(N,E,S,W) main kitna hain. 
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product',aggfunc='median') #SABKO EK SATH LANA, product ka konsa value(A,B,C,D) kis region(N,E,S,W) main kitna hain. 
pivot2
```
