```python
#SIMPLE MERGING AND CONCATENATION EXAMPLE

import pandas as pd
import numpy as np

data_employee = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'Diana Miller', 'Eve Davis'],
    'Department': ['HR', 'IT', 'Finance', 'HR', 'Marketing'],
    'Salary': [60000, 75000, 62000, 58000, 68000],
    'HireDate': ['2020-01-15', '2019-03-22', '2021-07-01', '2020-09-10', '2018-11-05']
}

df_employees = pd.DataFrame(data_employee)
df_employees
```

```python
data_products = {
    'ProductID': ['P001', 'P002', 'P003', 'P004', 'P005'],
    'ProductName': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Electronics', 'Peripherals'],
    'Jan_Sales': [120, 300, 250, 80, 150],
    'Feb_Sales': [130, 280, 260, 90, 140],
    'Mar_Sales': [110, 320, 240, 85, 160]
}

df_products = pd.DataFrame(data_products)
df_products
```

```python
# Concatenate the two DataFrames side by side
pd.concat([df_employees, df_products], axis=1)
```

```python
# Merge the two DataFrames on EmployeeID and ProductID
pd.merge(df_employees,df_products)
```
